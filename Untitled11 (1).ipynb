```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Dropout, LayerNormalization, MultiHeadAttention, Input, Flatten
from tensorflow.keras.optimizers import Adam
import tensorflow as tf

# Load data
ticker = 'AAPL'
data = yf.download(ticker, start='2020-01-01', end='2024-01-01')

# Preprocess data
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data[['Close']])

# Create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 60
X, y = create_sequences(data_scaled, seq_length)

# Split data
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Reshape target variable to avoid shape mismatch
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

# Build LSTM model
def build_lstm_model(input_shape):
    inputs = Input(shape=input_shape)
    x = LSTM(50, return_sequences=True)(inputs)
    x = Dropout(0.2)(x)
    x = LSTM(50)(x)
    outputs = Dense(1)(x)
    model = tf.keras.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='mse')
    return model

# Build GRU model
def build_gru_model(input_shape):
    inputs = Input(shape=input_shape)
    x = GRU(50, return_sequences=True)(inputs)
    x = Dropout(0.2)(x)
    x = GRU(50)(x)
    outputs = Dense(1)(x)
    model = tf.keras.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='mse')
    return model

# Build Transformer model
def build_transformer_model(input_shape):
    inputs = Input(shape=input_shape)
    x = LayerNormalization(epsilon=1e-6)(inputs)
    x = MultiHeadAttention(num_heads=4, key_dim=64)(x, x)
    x = Flatten()(x)
    x = Dense(50, activation='relu')(x)
    outputs = Dense(1)(x)
    model = tf.keras.Model(inputs, outputs)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    return model

# Train models
lstm_model = build_lstm_model((seq_length, 1))
lstm_model.fit(X_train, y_train, epochs=10, batch_size=32)

gru_model = build_gru_model((seq_length, 1))
gru_model.fit(X_train, y_train, epochs=10, batch_size=32)

transformer_model = build_transformer_model((seq_length, 1))
transformer_model.fit(X_train, y_train, epochs=10, batch_size=32)

# Predictions
y_pred_lstm = lstm_model.predict(X_test)
y_pred_gru = gru_model.predict(X_test)
y_pred_transformer = transformer_model.predict(X_test)

# Inverse transform results
y_pred_lstm = scaler.inverse_transform(y_pred_lstm)
y_pred_gru = scaler.inverse_transform(y_pred_gru)
y_pred_transformer = scaler.inverse_transform(y_pred_transformer)
y_test = scaler.inverse_transform(y_test)

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(y_test, label='Actual Prices', color='black')
plt.plot(y_pred_lstm, label='LSTM Predictions', color='blue')
plt.plot(y_pred_gru, label='GRU Predictions', color='green')
plt.plot(y_pred_transformer, label='Transformer Predictions', color='red')
plt.legend()
plt.show()
```
